```python
import math

def gaussian_reward(x, target_value, spread=1.0):
    """
    Calculate the reward based on a Gaussian function.
    The reward is 1 when x is equal to the target_value and decreases smoothly as x deviates from the target.
    """
    return math.exp(-(x - target_value)**2 / (2 * spread**2))

# Example usage:
target_value = 10.0  # Adjust this to your specific target value
current_value = 12.0  # Adjust this to your specific current value

reward = gaussian_reward(current_value, target_value, spread=1.0)
print(reward)
```

```python
rewards = []
for i in range(20):
    rew = gaussian_reward(i,10, spread=3.0)
    rewards.append(rew)

import matplotlib.pyplot as plt

plt.plot(rewards)
plt.show()
```
