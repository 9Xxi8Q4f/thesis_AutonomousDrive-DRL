```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
score = np.load("score_history.npy")
# score.reshape((1, len(score)))
average_score = 0
avg_history = []
for i in range(len(score)):
    average_score = np.mean(score[:i])
    avg_history.append(average_score)

plt.plot(avg_history)
```
